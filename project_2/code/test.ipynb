##### Test QFT implementation

```python
import sys

import numpy as np

sys.path.insert(0, "../../src/python")
from gate import fourier_transform
from qubit import NQubitState, random_qubit_state, basis_state
from fourier import qft_matrix

n_qubits = 3
qft = qft_matrix(n_qubits)  # QFT matrix for testing

# State vector
state = random_qubit_state(2**n_qubits)
# state = basis_state(n_qubits, 0)

# Apply Fourier transform
qubit = NQubitState(state)
# print(f"Initial state:\n{qubit.dump_state()}")

qubit.fourier_transform()
# print(f"Transformed state:\n{qubit.dump_state()}")

# Check if the transformed state is valid
test = qft @ state
print(f"Valid QFT state: {np.all(np.isclose(test, qubit.state))}")

# Check if the Fourier gate is valid
qft_gate = fourier_transform(n_qubits)
print(f"Valid QFT gate: {np.all(np.isclose(qft_gate, qft))}")
```

#### Quantum phase estimation

```python
import sys

import numpy as np

sys.path.insert(0, "../../src/python")
from qpe import qpe_test, random_unitary_matrix
from gate import Gate
from qubit import NQubitState

num_qubits = 3
precision_qubits = 6

u = Gate(random_unitary_matrix(num_qubits))

eigen_index = 0
eigenvalues, eigenvectors = np.linalg.eig(u)

phase = np.real(np.log(eigenvalues[eigen_index]) / (2.0j * np.pi))
if phase < 0:
  phase += 1.0

print(f"Phase: {phase:.3f}")
print(f"Eigenvalue: {eigenvalues[eigen_index]:.3f}")

eigenstate = NQubitState(eigenvectors[:, eigen_index])
phase_est, probability = qpe_test(u, eigenstate, precision_qubits)

print(f"Phase estimate: {phase_est:.3f} delta: {np.abs(phase_est - phase):.3f}")
print(f"Probability: {probability}")
```
